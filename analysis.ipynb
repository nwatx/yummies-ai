```python
import pandas as pd
```

```python
df = pd.read_csv("output.csv")
df
```

```python
import matplotlib.pyplot as plt
```

```python
df.columns
df = df.dropna(subset=["Rating"])
ratings = df['Rating']
ratings
```

```python
rating_num = []
for rating in ratings:
    split = rating.split(",")
    for n in split:
        rating_num.append(float(n))
        
rating_num = pd.Series(rating_num)
rating_num.describe()
```

```python
plt.hist(rating_num, bins=20)
plt.title("Rating Histogram")
plt.show()
```
